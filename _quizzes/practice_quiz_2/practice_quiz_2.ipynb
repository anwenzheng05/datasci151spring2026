**Q2** Simulate a sequence of dice rolls

<font size = "3">

What you did for Q1 is essentially how the `random.choice` function works:

``` python
    roll = random.choice([1,2,3,4])
```

(You can also use the NumPy library and use `numpy.random.choice` instead).

Suppose we want to see how many rolls it takes to roll a "3" **twice** (in total, not consecutively). Write code that simulates this experiment. After your code completes, the variable `num_rolls` should be equal to the number of rolls it took to roll the second "3".

**Example:**

If we roll the following numbers (reading left-to-right):

$$ 1, 4, 3, 1, 1, 1, 2, 3$$

then `num_rolls` should have the value 8.

````python
import random

num_rolls = 0
count_threes = 0

while count_threes < 2:
    roll = random.choice([1, 2, 3, 4])
    num_rolls += 1
    if roll == 3:
        count_threes += 1

print("Number of rolls to get two 3's:", num_rolls)
````

**Q3** Simulate a sequence of dice rolls (this one's a bit harder)

<font size = "3">

Suppose we want to see how many rolls it takes to roll a "3" **two consecutive times**. Write code that simulates this experiment. After your code completes, the variable `num_rolls` should be equal to the number of rolls it took to roll two "3's" in a row.

**Example:**

If we roll the following numbers (reading left-to-right):

$$ 1, 4, 3, 1, 1, 1, 2, 3, 2, 4, 3, 3$$

then `num_rolls` should have the value 12.

**Hint:** The computer needs to "remember" the previous dice roll. The most straightforward approach is to define a variable `previous_3 = False` and switch it to `True` when a 3 is rolled (don't forget to switch it back if you don't get two in a row!)

````python
import random

num_rolls = 0
consecutive_threes = 0

while consecutive_threes < 2:
    # Roll the die
    roll = random.choice([1, 2, 3, 4])
    num_rolls += 1
    
    # Update consecutive count
    if roll == 3:
        consecutive_threes += 1
    else:
        consecutive_threes = 0

print(f"It took {num_rolls} rolls to get two 3s in a row")
````

**Q4** Compute the sum of the first $n$ integers.

<font size = "3">

Using a `for` loop, compute the sum of all integers starting from 1 to the number stored in the variable $n$.

That is, compute the sum

$$ S_n = 1 + 2 + 3 + \cdots + n$$

````python
import random
n = random.randint(5, 100) # random number 5 <= n <= 100

# Compute the sum of integers from 1 to n
S_n = 0
for i in range(1, n + 1):
    S_n += i

print(f"n = {n}")
print(f"Sum of integers from 1 to {n} = {S_n}")
````

**Q5** Filter a data set

<font size = "3">

A temperature sensor in a warehouse records the temperature in Celsius every 30 minutes. Occasionally, the sensor "glitches" and return values that are unreasonable. In many applications, like this one, we often want to "clean" the data and remove unreasonable values before doing any analysis. 

Write code that loops over the list below. If any item is **outside** the range<br> $\textrm{min\_val} \leq \textrm{item} \leq \textrm{max\_val} $, replace it with the Python `None` object.

````python
temp_data = [-5, -2, 0, 1000, 2, -12, 44, 40, 30, -10]
min_val = -10
max_val = 40

# loop through the list by index so we can modify elements
for idx in range(len(temp_data)):
    val = temp_data[idx]
    # if value is strictly outside the allowed interval
    if val < min_val or val > max_val:
        temp_data[idx] = None

print(temp_data)
````

There are situations where we can't clean the data ahead of time, but still need to perform computations (e.g. we are writing software that uses the temperature data in real time as it is collected).

Using a `for` loop, compute the average of the readings in `temp_data` defined below, handling unreasonable values in one of two ways:

**Approach 1:** If an item is encountered that is **outside** the range $\textrm{min\_val} \leq \textrm{item} \leq \textrm{max\_val}$, ignore it. That is, compute the average of all the admissible values in the list. 

**Approach 2:** If an item is encountered that is **outside** the range $\textrm{min\_val} \leq \textrm{item} \leq \textrm{max\_val}$, print the message "Warning: error detected in sensor". Then return the average of the values that appeared **before** the error was detected.


````python
temp_data = [-5, -2, 0, 1000, 2, -12, 44, 40, 30, -10]
min_val = -10
max_val = 40

# here is what you should get for both approaches
approach_1_expected = (-5 + (-2) + 0 + 2 + 40 + 30 + (-10)) / 7

approach_2_expected = (-5 + (-2) + 0) / 3
````

````python
# Approach 1: Ignore values outside the range
temp_data = [-5, -2, 0, 1000, 2, -12, 44, 40, 30, -10]
min_val = -10
max_val = 40

sum_valid = 0
count_valid = 0

for value in temp_data:
    if min_val <= value <= max_val:
        sum_valid += value
        count_valid += 1

average_1 = sum_valid / count_valid
print(f"Approach 1 - Average: {average_1}")
print(f"Expected: {approach_1_expected}")
print(f"Match: {round(average_1, 10) == round(approach_1_expected, 10)}")
````

````python
# Approach 2: Stop at first error
temp_data = [-5, -2, 0, 1000, 2, -12, 44, 40, 30, -10]
min_val = -10
max_val = 40

sum_before_error = 0
count_before_error = 0

for value in temp_data:
    if value < min_val or value > max_val:
        print("Warning: error detected in sensor")
        break
    sum_before_error += value
    count_before_error += 1

print(f"Match: {round(average_2, 10) == round(approach_2_expected, 10)}")

average_2 = sum_before_error / count_before_errorprint(f"Expected: {approach_2_expected}")
print(f"Approach 2 - Average: {average_2}")
````
